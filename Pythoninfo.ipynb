```python
# AI-Assisted Threat Detection Dashboard

import pandas as pd
import matplotlib.pyplot as plt

# Sample network log data
data = {
    "User": ["Alice", "Bob", "Alice", "Charlie", "Bob", "David", "Alice"],
    "Failed_Logins": [1, 5, 2, 0, 7, 1, 8]
}

df = pd.DataFrame(data)

# Detect suspicious users
threshold = 5
df["Threat"] = df["Failed_Logins"].apply(
    lambda x: "Suspicious" if x >= threshold else "Normal"
)

print("Threat Detection Dashboard")
print(df)

# Plot
colors = ["red" if t == "Suspicious" else "green" for t in df["Threat"]]

plt.figure(figsize=(8,5))
plt.bar(df["User"], df["Failed_Logins"], color=colors)
plt.axhline(threshold, color="blue", linestyle="--", label="Threat Threshold")
plt.title("AI-Assisted Threat Detection Dashboard")
plt.xlabel("Users")
plt.ylabel("Failed Login Attempts")
plt.legend()
plt.show()
```
